```python
df.drop(['category', 'personName', 'source', 'countryOfCitizenship', 'organization', 'status', 'birthDate', 'title', 'date', 'state', 'residenceStateRegion',  'birthMonth', 'birthDay', 'tax_revenue_country_country', 'total_tax_rate_country', 'latitude_country', 'longitude_country'], axis =1, inplace = True)
```

```python
df.head()
```

```python
df.shape
```

```python
# To change the name of a column
df.rename(columns = {'age': 'Age', 'rank' : 'Rank', 'country': 'Country', 'city': 'City', 'industries': 'Industries', 'selfMade' : 'Self Made', 'gender': 'Gender', 'lastName':'Last Name', 'firstName' : 'First Name', 'birthYear': 'Birth Year','finalWorth' : 'NetWorth($B)', 'cpi_country' : 'CPI Index', 'cpi_change_country' : 'CPI Change', 'gdp_country' : 'Country GDP', 'gross_tertiary_education_enrollment' : 'Country tertiary education enrollment', 'gross_primary_education_enrollment_country' : 'Country primary education enrollment', 'life_expectancy_country' : 'Country life expectancy', 'population_country' : 'Population of country'}, inplace = True)
```

```python
df.head(10)
```

```python
# Convert networth value into billion 
df['NetWorth($B)'] = df['NetWorth($B)'] / 1000
```

```python
df = df.rename(columns=lambda x: x.strip())  # Remove extra spaces from all column names
df.dropna(subset=['Population of country'], inplace=True)
```

```python
df.dropna(subset=['Population of country'], inplace=True)
df['Population of country'] = df['Population of country'].astype(int)
```

```python
#find missing values in every column in the table
df.isnull().sum()
```

```python
df['Age'].value_counts()
```

```python
df.shape
```

```python
# use mean age to fill the 65 empty Age calues
mean_age = df['Age'].mean()
df['Age'].fillna(mean_age, inplace=True)
```

```python
# Replace blank strings values with Unknown
df['Country'].fillna('Unknown', inplace = True)
df['City'].fillna('Unknown', inplace = True)
df['Birth Year'].fillna('Unknown', inplace = True)
df['First Name'].fillna('Unknown', inplace = True)
```

```python
# Replace blank numeric values with Zero
df['CPI Index'] = df['CPI Index'].fillna(0)
df['CPI Change'] = df['CPI Change'].fillna(0)
df['Country GDP'] = df['Country GDP'].fillna(0)
df['Country tertiary education enrollment'] = df['Country tertiary education enrollment'].fillna(0)
df['Country primary education enrollment'] = df['Country primary education enrollment'].fillna(0)
df['Country life expectancy'] = df['Country life expectancy'].fillna(0)
```

```python
df['Population of country'].dtype
```

```python
df.head()
```

```python
# Remove $ and commas, then convert to integer
df['Country GDP'] = df['Country GDP'].str.replace('[\$,]', '', regex=True).astype('int64')
```

```python
df['Country primary education enrollment'] = (
    df['Country primary education enrollment'].astype(str) + '%'
)
```

```python
df['Country tertiary education enrollment'] = (
    df['Country tertiary education enrollment'].astype(str) + '%'
)
```

```python
#check for mixed data types 
for col in df.columns.tolist():
  weird = (df[[col]].map(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
# Convert to numeric, set errors='coerce' to turn invalid entries into NaN
df['Birth Year'] = pd.to_numeric(df['Birth Year'], errors='coerce')

# Now handle NaN (Option A or B above)
df['Birth Year'] = df['Birth Year'].fillna(0).astype(int)  # Example
```
